## GSE49756, GSE49755, GSE49757

Neutrophils as Reporter Cells: The study used neutrophils, a type of white blood cell important in the immune response, as "sensors" to detect immune signals in plasma (the liquid portion of blood containing proteins, antibodies, and signaling molecules).
Plasma Incubation:
Plasma from different groups (e.g., sepsis patients, healthy controls) was collected and added to neutrophils isolated from a healthy donor.
The neutrophils were cultured in vitro (in a controlled laboratory setting) with this plasma for 6 hours to allow interaction between plasma components and the neutrophils.
During this time, the plasma molecules (e.g., cytokines, immune proteins) triggered a transcriptional response (changes in gene expression) in the neutrophils.
The plasma samples were incubated with neutrophils for 6 hours, along with other conditions:
Medium alone (control).
Plasma from sepsis patients.
Plasma from uninfected controls.
Lipopolysaccharide (LPS) as a positive control for immune activation.

## GSE6535

Whole blood was taken from each patient on admission to ICU. Neutrophils were separated from whole blood using density-gradient separation with Ficoll-PaqueP P(Amersham).	The study included seventy-two critically ill patients admitted to the intensive care unit (ICU) of Nepean Hospital, Sydney, Australia. Of these, fifty-five patients were diagnosed to have sepsis, as confirmed by microbiological culture. The remaining seventeen patients did not have sepsis and were therefore used as controls. Critically ill patients with either gram-positive sepsis or gram-negative sepsis.
