## 21. Constructing a 10-core genes panel for diagnosis of pediatric sepsis" Journal of Clinical Laboratory Analysis (2021)(yellow,orange). 

Objective:

To construct and validate a diagnostic panel for pediatric sepsis using gene expression profiles derived from publicly available datasets.

Methods:

Gene Expression Analysis:
Downloaded three pediatric sepsis datasets (GSE13904, GSE25504, and GSE26440) from the GEO database.
Identified differentially expressed genes (DEGs) using the R package limma.
Conducted Weighted Gene Co-Expression Network Analysis (WGCNA) to identify sepsis-related modules.

Core Gene Selection:

Selected 10 hub genes (**HCK, PRKCD, SIRPA, DOK3, ITGAM, LTB4R, MAPK14, MALT1, NLRC3, LCK**) based on module analysis and verification using bioinformatics databases.

Panel Construction and Validation:

Constructed a diagnostic panel using elastic network regression.
Validated the panel in clinical datasets and patient samples.

Experimental Validation:
Quantified gene expression using qPCR in clinical samples of pediatric sepsis patients.

Key Findings:

Gene-Level Focus:The study focuses on gene expression, identifying genes differentially expressed in pediatric sepsis patients compared to healthy controls.
These genes were predominantly associated with immune response pathways, such as neutrophil-mediated immunity and cytokine signaling.

Diagnostic Panel:The 10-core gene panel demonstrated high diagnostic accuracy:
Area Under the Curve (AUC): 0.936.
Sensitivity: 96.6%.
Specificity: 58.7%.
Validation using independent datasets confirmed consistent performance with AUCs above 0.9, demonstrating robust diagnostic utility.

Biological Insights:Pathway analysis revealed that the core genes were enriched in pathways such as neutrophil degranulation, NF-κB signaling, and T-cell receptor signaling.
These findings highlight the relevance of immune system dysregulation in pediatric sepsis pathogenesis.

Diagnostic and Prognostic Value:

Diagnostic:The study emphasizes the diagnostic value of the 10-core gene panel, providing a tool for early identification of pediatric sepsis with high sensitivity.
Prognostic:
While primarily diagnostic, the panel may also aid in stratifying sepsis severity based on the immune response, although direct prognostic implications were not extensively discussed.

Conclusion:

The study demonstrates that a 10-core gene expression panel is a promising tool for diagnosing pediatric sepsis. It highlights the utility of integrating bioinformatics with clinical validation to identify biomarkers, focusing exclusively on gene expression rather than protein expression. The panel's high diagnostic accuracy underscores its potential for clinical application in managing pediatric sepsis.



## 22.T Cell– and Monocyte-Specific RNA-Sequencing Analysis in Septic and Nonseptic Critically Ill Patients and in Patients with Cancer

Objective:
To determine the effects of sepsis on immune cell subsets (CD4 T cells, CD8 T cells, and monocytes) by analyzing their gene expression profiles using RNA sequencing (RNA-seq). The study also compares immune responses in sepsis, nonseptic critical illness (CINS), and metastatic cancer.

Methods:

RNA-seq was conducted on purified CD4 T cells, CD8 T cells, and monocytes from:

29 septic patients.
23 critically ill nonseptic (CINS) patients.
5 cancer patients with metastatic colorectal tumors.
20 healthy controls.

Differential gene expression analysis identified immune pathways affected in each group.
Functional assays (e.g., ELISPOT) assessed immune responses, including T-cell cytokine production.

Key Findings:

Gene Expression Analysis:Sepsis caused significant immune suppression, with a focus on:

Monocytes:

Downregulation of immune response pathways (e.g., antigen presentation, TLR signaling).

Upregulation of immunosuppressive pathways, including IL-10 signaling.

CD4 T Cells:

Upregulation of cell cycle and stress-response pathways.
Increased expression of inhibitory molecules (e.g., PD-1, CTLA-4, LAG-3).

CD8 T Cells:

Less affected compared to monocytes and CD4 T cells, with fewer differentially expressed genes.

Comparative Analysis:Monocytes in sepsis, CINS, and cancer showed substantial overlap in immunosuppressive mechanisms.

CD8 T cells were more profoundly affected in cancer patients than in septic or CINS patients.
Common mechanisms included downregulation of HLA genes and upregulation of suppressor genes like SOCS1, SOCS3, and ARG1.

Functional Impairment:

Sepsis patients showed impaired T-cell responses, with reduced IFN-γ production, indicating defective adaptive immunity.
Monocyte HLA-DR expression was decreased in sepsis, consistent with suppressed antigen presentation.

Diagnostic and Prognostic Relevance:
Diagnostic:
Gene expression profiles identified immune suppression as a hallmark of sepsis, potentially aiding in distinguishing it from other conditions like cancer or nonseptic critical illness.

Prognostic:
The degree of immune suppression (e.g., HLA-DR downregulation, upregulation of inhibitory pathways) correlated with sepsis severity and might serve as a prognostic indicator for patient outcomes.

Conclusion:

The study highlights the gene-level analysis of immune suppression in sepsis, showing profound effects on monocytes and CD4 T cells. It underscores the potential of RNA-seq as a tool to uncover mechanisms of immunosuppression, which could inform both diagnostic and therapeutic strategies. Immune-based therapies targeting monocytes and T cells may hold promise for restoring immune competence in sepsis.

diagnostic markers for sepsis:

**HLA-DR**: Strong indicator of monocyte dysfunction and immune suppression in sepsis.

**PDCD1 (PD-1)**: Reflects T-cell exhaustion in septic patients.

**SOCS1/SOCS3**: Markers of cytokine signaling inhibition in monocytes.


## 23.Deciphering the Molecular and Cellular Atlas of Immune Cells in Septic Patients with Different Bacterial Infections

Objective:

To map the molecular and cellular profiles of immune cells in septic patients caused by different bacterial infections and to identify gene signatures and potential drug targets.

Methods:

Integrated bulk RNA sequencing, single-cell RNA sequencing (scRNA-seq), spatial transcriptomics, and metabolomics data from septic patients with 12 types of bacterial infections.
Identified signature genes for each bacterial infection.
Investigated immune cell heterogeneity, dysregulated pathways, and cell-to-cell communication in sepsis.
Performed drug screening and molecular optimization to identify targeted therapies.

Key Findings:

Gene-Level Analysis:Identified 19 signature genes specifically upregulated in E. coli sepsis, including DRAM1, which regulates autophagy and glycolysis in monocyte subsets.
DRAM1 was confirmed as a key gene involved in immune cell dysfunction in E. coli sepsis.

Cellular Heterogeneity:

Different types of bacterial sepsis were associated with distinct immune cell clusters:

E. coli sepsis: Linked to monocyte clusters (e.g., C10_ULK1) characterized by high autophagy and inflammatory activity.

S. viridans sepsis: Associated with specific monocyte and macrophage clusters.

Neisseria sepsis: Involved phagocytosis-related clusters enriched with C1QA gene expression.

Functional Pathways:

Upregulated pathways in E. coli sepsis included:
Autophagy.
Glycolysis.
Inflammatory signaling (e.g., resistin signaling).
Common pathways across sepsis types included neutrophil degranulation, innate immune response, and complement activation.

Potential Drug Targets:

Identified retapamulin as a drug targeting DRAM1 for E. coli sepsis.

Additional drugs were proposed for other sepsis types, including:
Nystatin for Neisseria sepsis (targeting C1QA).
Dalfopristin for S. viridans sepsis (targeting CTSD).

Spatial Transcriptomics in Kidney Injury:Acute kidney injury in sepsis involved cell-cell interactions mediated by integrin signaling.
Monocyte subsets (e.g., C10_ULK1) interacted with kidney cells, exacerbating inflammation and damage.

Diagnostic and Prognostic Relevance:

Diagnostic:The identified signature genes, such as **DRAM1 in E. coli sepsis**, provide a basis for distinguishing sepsis types and understanding their molecular mechanisms.
Prognostic:
Dysregulated pathways and immune cell interactions, especially involving monocytes, may serve as prognostic indicators of sepsis severity and organ dysfunction.

Conclusion:
This comprehensive study integrates multi-omics approaches to reveal the molecular and cellular heterogeneity in sepsis caused by different bacterial infections. Signature genes like DRAM1 are promising targets for stratified therapy, and the study highlights the importance of personalized treatment strategies in managing sepsis.

Upregulated genes: **C1QA, CTSD, DRAM1, ULK1, SOCS1, SOCS3**.

Downregulated genes: **HLA-DRA, HLA-DRB1**.


## 24.integrative Analysis of Metabolomics and Transcriptomics to Uncover Biomarkers in Sepsis

 Objective: 
 
 To identify novel biomarkers for diagnosing and managing sepsis by combining transcriptomics and metabolomics data from septic patients.
 
Methods:

Blood samples were collected from 16 sepsis patients and 11 systemic inflammatory response syndrome (SIRS) patients.
RNA sequencing and non-targeted metabolomics (using LC–MS/MS) were performed.
Integration of transcriptomics and metabolomics data helped identify co-regulated biomarkers and pathways.

Key Findings:

Gene Expression Analysis:
485 differentially expressed genes were identified:
301 upregulated.
184 downregulated.
Functional enrichment revealed involvement in inflammatory response, immune regulation, and amino acid metabolism.

Metabolomics Integration:
1083 metabolites showed differential expression:
831 upregulated.
252 downregulated.
Combined analysis highlighted pathways like glutathione metabolism, mucin-type O-glycan biosynthesis, and cysteine/methionine metabolism.

Key Diagnostic Genes:

The meta-analysis identified four core genes with significant upregulation in sepsis compared to SIRS:

ITGAM: Encodes integrin CD11b, involved in macrophage adhesion and immune regulation.

CD44: Plays a role in cell adhesion and immune cell migration.

C3AR1: A receptor for complement component 3a, regulates inflammation.

IL2RG: Common gamma chain of cytokine receptors, critical for immune signaling.

Single-Cell RNA Sequencing:

**ITGAM and C3AR1** were predominantly expressed in macrophage clusters.
CD44 and IL2RG were expressed across various immune cell types, reflecting their widespread involvement in immune modulation.

Pathway Analysis:

Highlighted pathways included:
Th1/Th2 cell differentiation.
Staphylococcus aureus infection pathways.
Hematopoietic cell lineage.

Diagnostic and Prognostic Relevance:
Diagnostic:
The core genes (**ITGAM, CD44, C3AR1, IL2RG**) provide potential biomarkers for distinguishing sepsis from SIRS and tracking disease progression.

Prognostic:
The expression levels of ITGAM and C3AR1 in macrophages correlate with inflammatory responses, suggesting their use as markers for immune dysregulation severity.

Conclusion:
The study identifies ITGAM, CD44, C3AR1, and IL2RG as promising diagnostic biomarkers for sepsis. The integration of metabolomics and transcriptomics data provides a comprehensive understanding of sepsis pathophysiology and highlights key molecular targets for further research and therapeutic intervention.
