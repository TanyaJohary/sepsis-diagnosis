## 15. Characterising Cytokine Gene Expression Signatures in Patients with Severe Sepsis
Journal: Mediators of Inflammation, 2013.

Objectives:

Investigate the differences in cytokine gene expression profiles between:
Healthy individuals.
Patients with infections without organ failure.
Patients with severe sepsis.
Develop a diagnostic model for distinguishing severe sepsis from infection based on cytokine gene expression.

Methods:

Prospective observational study conducted on:

54 patients with severe sepsis.
53 patients with infection but no organ failure.
20 healthy controls.
Blood samples were collected from all participants, and PBMCs (Peripheral Blood Mononuclear Cells) were isolated.

Gene Expression Analysis:

Absolute quantification of cytokine mRNA levels using RT-qPCR.

Genes analyzed: TNF-α, IFN-γ, IL-2, IL-7, IL-10, IL-23, IL-27.

Cytokine protein levels were assessed for IL-6 using ELISA.

Statistical Analysis:

Wilcoxon rank-sum and Kruskal-Wallis tests were used for group comparisons.
Logistic regression was performed to identify significant predictors of sepsis.

Key Findings:

Gene Expression Profiles:

Compared to Healthy Controls:
Patients with infection had:
Decreased expression of IL-2 and IL-23.
Increased expression of IL-10 and IL-27.

Compared to Infection:
Patients with severe sepsis had:

Lower expression of IL-7, IL-23, IFN-γ, and TNF-α.
Higher expression of IL-10.
IL-27 levels remained unchanged between these groups.

Diagnostic Index:

A cytokine index was developed using the formula:
(IFN-γ + TNF-α + IL-7 + IL-23) - IL-10.
This index had:
AUC (Area Under Curve): 0.88.
Sensitivity: 90%.
Specificity: 83%.
Positive predictive value: 81%.
Negative predictive value: 90%.

Correlation with Disease Severity:

Serum IL-6 levels correlated significantly with organ failure severity (SOFA score) but not with cytokine gene expression.

Cytokine gene expression profiles did not change significantly over time in severe sepsis patients.

Conclusion:

Immune Dysregulation:

Severe sepsis is characterized by a unique gene expression profile involving downregulation of pro-inflammatory and homeostatic cytokines (TNF-α, IFN-γ, IL-7) and upregulation of the anti-inflammatory cytokine IL-10.

Diagnostic Potential:Cytokine gene expression profiling can serve as a robust diagnostic approach to distinguish sepsis from infection.

No Prognostic Value:Cytokine expression was not correlated with mortality or organ failure severity.

Significance:

This study provides a molecular understanding of immune regulation in sepsis and suggests cytokine mRNA profiling as a potential diagnostic biomarker for sepsis.

The diagnostic genes identified in this study as part of the cytokine gene expression signature for distinguishing severe sepsis from infection are:

Diagnostic Genes:

Pro-inflammatory and Homeostatic Genes (Downregulated in Severe Sepsis):

**IFN-γ** (Interferon-gamma): Plays a key role in activating macrophages and promoting the Th1 immune response.

**TNF-α** (Tumor Necrosis Factor-alpha): A major pro-inflammatory cytokine driving the acute inflammatory response.

**IL-7** (Interleukin-7): Essential for T-cell homeostasis and immune recovery.

**IL-23** (Interleukin-23): Promotes Th17 responses and inflammation.

Anti-inflammatory Gene (Upregulated in Severe Sepsis):

**IL-10 (Interleukin-10)**: Suppresses pro-inflammatory responses and is associated with immune suppression during sepsis.

The upregulation of IL-10 and the downregulation of IFN-γ, TNF-α, IL-7, and IL-23 create a unique expression pattern in severe sepsis.
These genes can be used as a diagnostic signature to differentiate sepsis from infection without organ failure.




## 16. Sepsis Upregulates the Gene Expression of Multiple Ubiquitin Ligases in Skeletal Muscle
Journal: The International Journal of Biochemistry & Cell Biology, 2003.(not found)

Objectives:

Investigate the effects of sepsis on the expression of ubiquitin ligases in skeletal muscle.

Evaluate the role of glucocorticoids in the regulation of these ligases during sepsis.

Explore the differential expression of ubiquitin ligases in fast-twitch (extensor digitorum longus, EDL) and slow-twitch (soleus) muscles.

Animal Model:

Male Sprague–Dawley rats (40–60 g) were used.

Sepsis was induced using the cecal ligation and puncture (CLP) method. Sham-operated rats served as controls.
In some groups, the glucocorticoid receptor antagonist RU 38486 was administered to assess the role of glucocorticoids.

Gene Expression Analysis:

Real-time PCR was used to measure the mRNA levels of:

MuRF1 (Muscle Ring Finger 1).

MAFbx (Muscle Atrophy F-box)/atrogin-1.

The abundance of mRNA was compared between septic and sham-operated rats.

Tissue-Specific Analysis:
The study focused on the EDL (fast-twitch) and soleus (slow-twitch) muscles.

Key Findings:

Increased Gene Expression:

Sepsis caused a significant upregulation of MuRF1 and MAFbx/atrogin-1 mRNA in the EDL muscle:

**MuRF1**: 10-fold increase.

**MAFbx**: 12-fold increase.

**Atrogin-1**: 16-fold increase.

These changes were much greater than the 2-fold increase observed for the ubiquitin ligase E3α in previous studies.

Muscle-Specific Effects:

The upregulation of MuRF1 and MAFbx/atrogin-1 was observed only in the fast-twitch EDL muscle.
No significant changes were detected in the slow-twitch soleus muscle or liver tissue.

Role of Glucocorticoids:

Administration of RU 38486 (glucocorticoid receptor antagonist) prevented the sepsis-induced increase in MuRF1 and MAFbx/atrogin-1 mRNA levels.
This indicates that glucocorticoids are critical mediators of ubiquitin ligase upregulation during sepsis.

Clinical Implications:
The upregulation of these ubiquitin ligases suggests their involvement in sepsis-induced muscle wasting.
These ligases may serve as potential biomarkers or therapeutic targets for managing muscle atrophy in sepsis.

Discussion:

Mechanisms of Muscle Wasting:
Ubiquitin ligases like MuRF1 and MAFbx are key regulators of the ubiquitin-proteasome pathway, which mediates protein degradation in muscle.
The study suggests that multiple ligases are involved in targeting different proteins or classes of proteins for degradation during sepsis.

Differential Regulation:
The upregulation of ubiquitin ligases in fast-twitch muscles highlights tissue-specific responses to sepsis, possibly due to differences in metabolism and protein turnover rates.

Conclusions:

Sepsis significantly increases the expression of MuRF1 and MAFbx/atrogin-1 in skeletal muscle, driven largely by glucocorticoid signaling.
These ligases may serve as sensitive markers for sepsis-induced muscle wasting and potential targets for therapeutic intervention.


## 17. Hypoxia-inducible Factor and Target Gene Expression Are Decreased in Patients with Sepsis
Journal: Anesthesiology (2013).

Objectives:

Investigate the expression of Hypoxia-inducible factor-1 alpha (HIF-1α) in patients with sepsis compared to healthy individuals.

Assess the expression of HIF-1α target genes, including adrenomedullin, CD11a, CD11b, and CD18.

Study the effect of lipopolysaccharide (LPS) on HIF-1α expression in monocytic cells.

Methods:

Subjects: 99 patients with sepsis and 48 healthy volunteers.
Blood samples were collected, and leukocyte HIF-1α expression was analyzed using RT-PCR and immunofluorescence.

In Vitro Experiments:

THP-1 monocytic cells were treated with LPS to assess the effect of acute and repetitive stimulation on HIF-1α mRNA and protein expression.

Analysis:

Cytokine concentrations (e.g., IL-6, IL-10, TNF-α) were measured.
Statistical analysis was performed to correlate HIF-1α expression with disease severity (SAPS II scores) and clinical outcomes.

Key Findings:

HIF-1α Expression:HIF-1α mRNA was significantly decreased in sepsis patients compared to healthy controls (67% reduction, P = 0.0001).
The reduction was inversely correlated with illness severity (r = -0.29, P = 0.0001).
HIF-1α protein levels were similarly reduced in sepsis patients.

Target Gene Expression:

**Adrenomedullin**, **CD11a**, and **CD11b** expression were reduced by 75–86% in sepsis patients (P = 0.0001).

CD18, a common β2-integrin subunit, showed a slight increase in expression.

Cytokine Levels:

IL-6, IL-10, and TNF-α levels were significantly elevated in sepsis patients, reflecting an inflammatory response.

IFN-α and IFN-γ levels were not significantly altered.

In Vitro Results:

Acute LPS stimulation increased HIF-1α mRNA and protein expression in naive THP-1 cells.
Repetitive LPS stimulation led to endotoxin tolerance, reducing HIF-1α mRNA and protein levels.

Clinical Implications:
Reduced HIF-1α expression was linked to impaired activation of genes critical for immune response and tissue oxygenation.

Conclusions:

HIF-1α and its target gene expressions are significantly suppressed in sepsis, correlating with disease severity.
Decreased HIF-1α levels may impair immune function and adaptation to hypoxia in septic patients.
These findings highlight HIF-1α as a potential biomarker and therapeutic target in sepsis.

Diagnostic Role:

**HIF-1α** expression and its downstream targets (e.g., adrenomedullin, CD11a, CD11b) could serve as diagnostic biomarkers for sepsis severity and immune dysregulation.

## 18. Altered Melatonin Secretion and Circadian Gene Expression in Early-Stage Sepsis Patients
Journal: Molecular Medicine Reports, 2013.

Objectives:

Investigate the circadian rhythm of melatonin secretion in patients with early-stage sepsis compared to non-septic ICU patients.

Evaluate the expression of circadian genes Cry-1 and Per-2 in peripheral blood leukocytes (PBLs) from these patients.

Examine the relationship between altered melatonin rhythms and the secretion of pro-inflammatory cytokines TNF-α and IL-6.

Methods:

22 ICU patients divided into:
Septic group: 11 patients with early-stage sepsis.
Non-septic group: 11 ICU patients without sepsis.
Blood samples were collected every 4 hours over a 24-hour period.

Analysis:

Melatonin Levels:Plasma melatonin was measured using radioimmunoassay.

Gene Expression:mRNA levels of Cry-1 and Per-2 were quantified using RT-qPCR.

Cytokine Levels:Plasma TNF-α and IL-6 were measured using ELISA.

Statistical Approach:Comparisons were made using the Mann-Whitney U test or Student’s t-test.
Significance was defined as P < 0.05.

Key Findings:

Altered Melatonin Rhythm:

In sepsis patients:Peak melatonin secretion occurred at 6:00 p.m., significantly earlier than the typical 2:00 a.m. peak in non-septic patients.
Although melatonin levels were generally higher in the septic group, this difference was not statistically significant (P = 0.055).

Circadian Gene Expression:mRNA levels of Cry-1 and Per-2 were significantly reduced in sepsis patients compared to non-septic controls:
Cry-1: 2.10 ± 0.32 (sepsis) vs. 4.30 ± 0.66 (non-sepsis), P < 0.001.
Per-2: 0.78 ± 0.12 (sepsis) vs. 1.09 ± 0.19, P = 0.01.

Cytokine Levels:Plasma levels of TNF-α and IL-6 were significantly elevated in septic patients:
TNF-α: 64.1 ng/mL (sepsis) vs. 11.4 ng/mL (non-sepsis), P = 0.04.
IL-6: 41.2 ng/mL (sepsis) vs. 19.1 ng/mL, P = 0.04.
Peak cytokine levels coincided with the shifted melatonin secretion peak at 6:00 p.m.

Clinical Characteristics:Sepsis patients had higher SOFA scores (10.8 ± 6.13 vs. 6.45 ± 2.84, P = 0.045) and lower mean arterial pressure (71.82 ± 24.96 mmHg vs. 109.27 ± 30.96 mmHg, P = 0.006) compared to non-septic patients.

Discussion:

Melatonin and Sepsis:The altered circadian rhythm of melatonin secretion in sepsis patients reflects disrupted physiological regulation.
Elevated TNF-α and IL-6 levels suggest an inflammatory influence on melatonin production.

Circadian Gene Dysregulation:Suppression of Cry-1 and Per-2 indicates a breakdown of circadian regulation in peripheral tissues during early-stage sepsis.

Interplay of Melatonin and Inflammation:
The study suggests a possible feedback loop where inflammation alters circadian rhythms, and disrupted melatonin secretion exacerbates immune dysregulation.

Conclusions:

The circadian rhythm of melatonin secretion and circadian gene expression is altered in early-stage sepsis.
**Cry-1**, **Per-2** gene, and melatonin levels may have potential as biomarkers for early sepsis diagnosis.
Further studies are needed to clarify the mechanisms linking circadian disruptions and immune responses in sepsis.

## 19. Differential Gene Expression in Circulating CD14+ Monocytes Indicates the Prognosis of Critically Ill Patients with Sepsis
Journal: Journal of Clinical Medicine, 2020.

Objectives:

Investigate the gene expression profiles of CD14+ monocytes in critically ill patients, with and without sepsis.
Identify specific genes associated with sepsis severity and mortality.
Evaluate the potential of these genes as prognostic markers in sepsis.

Methods:

Patient Cohorts:

54 healthy controls.
42 standard care (SC) patients with bacterial infections.
76 ICU patients (36 with sepsis, 40 without sepsis).
Longitudinal samples were collected on days 1, 3, and 7 during ICU stays.
Diagnostic criteria for sepsis followed SOFA score guidelines.

Gene Expression Analysis:

RNA sequencing of CD14+ monocytes for an unbiased approach.
NanoString technology used to validate expression of 55 highly variable genes.

Flow Cytometry:
Characterization of monocyte subsets (classical, intermediate, and non-classical) and their surface markers (e.g., HLA-DR, CX3CR1).

Statistical Methods:
Principal Component Analysis (PCA).
Receiver Operating Characteristic (ROC) curve analysis for prognostic markers.
Kaplan–Meier survival analysis.

Key Findings:

Gene Expression in Sepsis:

Significant differences in the expression of 55 genes in CD14+ monocytes were identified between septic and non-septic patients.

Two genes stood out:

**ALOX5AP** (Arachidonate 5-Lipoxygenase-Activating Protein):Upregulated in sepsis patients compared to non-sepsis and healthy controls.Associated with the severity of sepsis but showed limited predictive power for mortality.

**ARHGEF10L** (Monocytic Guanine Nucleotide Exchange Factor 10-Like Protein):
Downregulated in sepsis patients.
Strongly associated with ICU mortality and survival outcomes.

Monocyte Subsets:

Classical monocytes were reduced in sepsis patients, while intermediate and non-classical monocytes were elevated.
HLA-DR and CX3CR1 expression was reduced across monocyte subsets in sepsis patients, correlating with poor prognosis.

Prognostic Value:

ARHGEF10L demonstrated significant prognostic value:
Lower expression was associated with higher mortality.
Kaplan–Meier analysis revealed higher survival in patients with higher ARHGEF10L expression.
ROC analysis yielded an AUC of 0.73 for ARHGEF10L as a prognostic marker.

Longitudinal Observations:In septic patients, ALOX5AP remained persistently upregulated, and ARHGEF10L downregulated throughout the ICU stay.
Non-septic patients showed normalization of gene expression over time, reflecting recovery.

Conclusions:

ALOX5AP and ARHGEF10L are potential markers for monitoring disease progression and severity in sepsis.
ARHGEF10L is particularly valuable as a prognostic biomarker, given its association with patient survival.
The study emphasizes the importance of monocytic gene expression profiles in understanding and managing sepsis.

Clinical Implications:

These findings could aid in the development of targeted therapies and improved patient stratification based on monocyte activation profiles.

ALOX5AP (Arachidonate 5-Lipoxygenase-Activating Protein):

Expression: Upregulated in sepsis patients compared to non-sepsis and healthy controls.
Function:Involved in the production of leukotrienes, which are inflammatory mediators.
Plays a role in the immune response and inflammation, particularly in the context of bacterial infections.
Clinical Relevance:Associated with the severity of sepsis but not strongly predictive of mortality.

ARHGEF10L (Rho Guanine Nucleotide Exchange Factor 10-Like Protein):
Expression: Downregulated in sepsis patients, particularly in those who did not survive.
Function:Plays a role in cell signaling and cytoskeletal organization.
Important for maintaining monocyte functions during immune responses.
Clinical Relevance:Strongly correlated with patient outcomes, making it a potential prognostic marker for sepsis mortality.Lower levels were associated with higher mortality rates.
HLA-DR and CX3CR1 were evaluated at the protein level in this study.


## 20. Patterns of Gene Expression in Peripheral Blood Mononuclear Cells and Outcomes from Patients with Sepsis Secondary to Community Acquired Pneumonia
Journal: PLOS ONE, 2014.

Objective:

To analyze the gene expression profiles in peripheral blood mononuclear cells (PBMCs) from patients with sepsis caused by community-acquired pneumonia (CAP) and compare:
Survivors vs. non-survivors of sepsis.
Early-stage (diagnosis) and later-stage (day 7) samples to evaluate their relation to clinical outcomes.

Methods:

10 patients with severe sepsis or septic shock caused by CAP (5 survivors, 5 non-survivors).
3 healthy volunteers served as controls.

Sample Collection:

Blood samples collected:
At sepsis diagnosis (D0).
7 days after diagnosis (D7).

Gene Expression Analysis:

PBMCs were isolated, RNA extracted, and analyzed using Agilent Whole Human Genome Microarrays.
Principal Component Analysis (PCA) and Gene Ontology (GO) term enrichment analysis were performed to identify patterns in gene expression.

Statistical Analysis:

Genes with a fold change ≥ 1.7 and a P-value ≤ 0.05 were classified as differentially expressed.
KEGG pathway analysis was used to map biological pathways.

Key Results:

Gene Expression in Septic Patients vs. Controls:

151 genes were differentially expressed:
104 upregulated (e.g., TLR2, TLR4, TNF, IL1B, S100A12).
49 downregulated.
PCA clustered septic patients distinctly from healthy controls, with survivors and non-survivors forming separate subgroups.

Immune Response Genes:

Upregulated genes were enriched in pathways associated with:
Toll-like receptor (TLR) signaling (e.g., TLR2, TLR4).
Inflammatory response (e.g., IL8, TNF, NFKBIZ).
Chemotaxis (e.g., CXCL2, CCR1).

Survivors vs. Non-Survivors:

D0 Samples:
28 genes involved in energy metabolism (e.g., mitochondrial electron transport chain) were upregulated in survivors but not in non-survivors.

D7 Samples:
Non-survivors showed decreased expression of immune-related genes (e.g., HLA-DRB1, CD74) compared to survivors.
Survivors maintained higher levels of immune function-related gene expression.

Oxidative Phosphorylation:
Survivors exhibited higher expression of genes related to the mitochondrial electron transport chain, suggesting better energy metabolism compared to non-survivors.

Disease Progression:
PCA showed that gene expression patterns observed at diagnosis persisted over time, particularly distinguishing survivors from non-survivors.

Conclusions:

Diagnostic Implications:
Upregulated genes in sepsis (e.g., TLR2, TLR4, IL8, TNF) reflect the activation of immune pathways critical for fighting infection.
These patterns distinguish sepsis patients from healthy controls but were not directly tested as diagnostic markers.

Prognostic Implications:
Survivors maintained higher expression of immune response and oxidative phosphorylation genes, while non-survivors showed immune suppression over time.
These findings suggest the potential of immune function-related genes (e.g., HLA-DRB1, TLR4) as prognostic biomarkers.

Level of Expression:
The study exclusively analyzed gene-level expression (mRNA) using microarray technology.


Differentially Expressed Genes:

Up-regulated Genes in Sepsis:

Immune response: **TLR2, TLR4, IL1B, TNF, S100A12, IL8**.

Chemotaxis: **CXCL2, CXCL3, CCR1**.

Antigen presentation: **HLA-DRB1, HLA-DRA, CD74**.

Down-regulated Genes in Non-Survivors:

Mitochondrial function: **ATP5B, COX5A, COX10**.

Zinc transport: **SLC39A6, SLC39A9, SLC39A11**.
