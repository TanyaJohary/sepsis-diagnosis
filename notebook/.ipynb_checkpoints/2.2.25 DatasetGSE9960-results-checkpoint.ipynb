# Dataset: GSE9960

SE9960 is a publicly available dataset in the Gene Expression Omnibus (GEO) repository, titled "Gene-expression profiling of peripheral blood mononuclear cells in sepsis." The study aimed to identify signature genes that help distinguish sepsis from non-infectious causes of systemic inflammatory response syndrome and to differentiate between Gram-positive and Gram-negative sepsis.

## Design: 

dataset include sepsis by gram negative and gram positive and other infectious pathogens and control one. We considered all kinds of infection as sepsis and the control greop as control as well.

The dataset was already normalized by RMA and Quantile normalization. Because the dataset is imbalanced, we used the SMOTE resampling method.

### Feature removal results:

Each feature has been removed from the dataset and all metrics values are calculated to find out the most important genes in true prediction. (feature_removal_results_SMOTE_GSE9960.csv)


Genes with High Impact (MCC drops significantly):(**TLR4.1, ARG1, MYD88,MAPK14.2, BCL2**)
Removing these genes causes a sharp decline in MCC, indicating they are critical for the model's classification performance.

These are likely the most informative genes for predicting the outcome (e.g., sepsis).

Feature importance results:
Feature importance extraction in Random Forest helps us understand which features (genes, biomarkers, etc.) contribute the most to classification performance. we this for each iteration and then make average of all the results.(average_feature_importance_SMOTE_GSE9960.csv)



Based on the results of feature extraction we can see some genes such as (**MMP8.1,MAPK14.2, MAPK14.1, S100A8, SOCS3**,) are the most influential ones in classification of sepsis and healthy patients. So these features reduce impurity (uncertainty) when used in tree splits and with higher Gini scores are more important.


## Sanity Check:

This plot represents the sanity check where different metrics are tracked as noise levels increase, with a dataset balanced using SMOTE (Synthetic Minority Oversampling Technique).(sanity_check_results_GSE9960.csv)


Higher noise levels negatively impact model performance.

AUC is the most robust to noise, while MCC is the most affected.

The F1 Score shows a moderate decline, suggesting that both false positives and false negatives are increasing as noise is introduced.

## Mann-Whitney U test:

The Mann-Whitney U test (also called the Wilcoxon rank-sum test) is a non-parametric test used to compare two independent groups when the assumptions of normality (required for a t-test) are not met. In our analysis, we want to compare gene expression between sepsis patients and controls. The Mann-Whitney test helps determine if the expression levels are significantly different between these two groups.(Man-W-GSE9960code.R)

If the p-value < 0.05, it means that there is a significant difference in gene expression between Sepsis vs. Control for that gene.

If p-value â‰¥ 0.05, it means there is no strong evidence of a difference in expression levels.

We can see that gene MAPK14 has significantly different expression levels between sepsis and Control ones.
