#### 12.Bench-to-bedside review: Future novel diagnostics for sepsis – a systems biology approach
published in Critical Care

1. Genomics

Genomics involves studying genetic predisposition to sepsis using techniques like SNP analysis and GWAS.
Biomarkers and Genes:
SNPs (Single Nucleotide Polymorphisms):
**SNPs in IL-2 (Interleukin-2) gene**: Associated with increased susceptibility to bacteremia, malaria, and tuberculosis.
**TNF (Tumor Necrosis Factor) gene**:SNP at position -308 in the promoter region associated with higher TNFα expression, leading to increased mortality in septic shock.
**TLR4 (Toll-Like Receptor 4)**:SNPs linked to impaired immune response and gram-negative septic shock susceptibility.
**CD14**: SNPs in the CD14 locus are associated with susceptibility to septic shock and higher mortality.
Protein C Gene (-1641 AA Polymorphism): Linked to decreased survival, increased organ dysfunction, and systemic inflammation in severe sepsis.
Diagnostic Value:
SNPs offer insights into genetic predisposition but have limited diagnostic utility without further validation.
Useful for risk stratification and designing personalized therapeutic approaches.

2. Epigenomics

Epigenomics studies reversible DNA modifications (e.g., methylation and histone modifications) that influence gene activity.
Biomarkers and Genes:
**NF-κB (Nuclear Factor Kappa B)**:Epigenetically regulated through histone phosphorylation and acetylation.
Central to cytokine production in sepsis, including IL-6.

DNA Methylation in CpG Islands:Promoters of cytokine-related genes show altered methylation patterns affecting transcription.
Diagnostic Value:
Epigenetic markers are promising but require tissue-specific studies and tools for clinical application.

3. Transcriptomics

Transcriptomics quantifies gene expression levels through mRNA and miRNA profiling.
Biomarkers and Genes:

mRNA Expression:
28-gene signature: Identified as upregulated in survivors versus non-survivors of septic shock.
Altered expression in over 500 genes in septic patients, including cytokine and signal transduction genes.

miRNA Expression:
**miR-150**: Significantly downregulated in sepsis and correlates with organ failure scores.
**miR-146b**: Downregulated during endotoxemia.
**miR-143**: Upregulated in response to endotoxin.
**miR-342**: Downregulated in immune cells during sepsis.
**let-7g**: Downregulated in sepsis patients.
Diagnostic Value:
miR-150 is a strong prognostic marker for disease severity.
mRNA signatures are useful for early diagnosis but lack consistency across studies.

4. Proteomics

Proteomics focuses on protein profiling and discovery.
Biomarkers and Genes:
Known Biomarkers:

**Procalcitonin (PCT)**: An established marker for bacterial infections.
**C-Reactive Protein (CRP)**: A general marker of inflammation.
Novel Proteins Identified:
Plasma proteins with differential expression between survivors and non-survivors.
Cytokines and other uncharacterized proteins.
Diagnostic Value:
PCT and CRP are routinely used but lack specificity for sepsis.
Emerging proteins show promise for stratifying patients but require further study.

5. Metabolomics

Metabolomics studies small molecules (metabolites) involved in cellular processes.
Biomarkers:
**Glutathione**: Identified in sepsis-induced acute lung injury.
**Adenosine**: Altered levels in sepsis.
**Phosphatidylserine**: Changes associated with inflammatory states.
**Sphingomyelin**: Distinct levels observed in sepsis patients versus healthy controls.
Diagnostic Value:
Metabolomics biomarkers reflect downstream metabolic changes in sepsis.
Promising for disease progression monitoring and therapeutic response prediction.

Conclusion
The article identifies numerous biomarkers across genomics, epigenomics, transcriptomics, proteomics, and metabolomics. While many biomarkers show potential for diagnosis and prognosis, their clinical application remains limited without large-scale validation studies. Combining these markers into multi-omics panels offers the greatest promise for improving sepsis diagnosis, risk stratification, and personalized treatment.

#### 14.Early Diagnosis of Sepsis: Is an Integrated Omics Approach the Way Forward?
published in Molecular Diagnosis & Therapy in 2017

1. Current Challenges in Sepsis Diagnosis
Existing biomarkers like lactate lack specificity.
Clinical definitions such as Sepsis-3 criteria are not predictive of patient outcomes.
Misdiagnosis can lead to overtreatment with antimicrobials, contributing to antimicrobial resistance.

2. Multi-Omics Integration

A. Genomics
Genomic studies identify single nucleotide polymorphisms (SNPs) that influence sepsis susceptibility and outcomes.
Genes Mentioned:

**TLR4 (Toll-Like Receptor 4)**: Associated with immune response dysregulation and susceptibility to gram-negative infections.

**TLR2 (Toll-Like Receptor 2)**: Similar involvement in pathogen recognition.

**CCR5 (C-C Chemokine Receptor Type 5)**: Implicated in immune modulation.

**CFP (Complement Factor Properdin)**: Linked to complement activation and infection severity.

**SLC11A1 (Solute Carrier Family 11 Member A1)**: Associated with intracellular pathogen resistance.

B. Transcriptomics
Transcriptomics evaluates changes in gene expression to predict sepsis outcomes and differentiate between bacterial and viral infections.

Key Transcriptomic Findings:
Host Gene Expression Signatures: Differentiate between sepsis survivors and non-survivors.

miRNAs:

**miR-223**: Anti-inflammatory; protects against excessive immune activation.
**miR-146a**: Regulates inflammatory signaling pathways.

C. Proteomics

Proteomics identifies proteins that are upregulated or downregulated during sepsis.
Proteins Identified:

**Cardiolipin**: Mitochondrial membrane component linked to organ dysfunction.

**Mitochondrial DNA (mtDNA)**: Released during cell damage; acts as a damage-associated molecular pattern (DAMP).

D. Metabolomics

Metabolomics evaluates biochemical changes in small molecules.
Key Metabolites:

**Carnitine Esters**: Indicators of mitochondrial dysfunction.

**Bile Acids**: Elevated in non-survivors.

**Nicotinamide Adenine Dinucleotide (NAD+)**: Associated with energy metabolism disruptions.

**Tricarboxylic Acid (TCA) Cycle Intermediates:** Suggest an energy crisis in non-survivors.

3. Diagnostic Value of Biomarkers

**Lactate**: Still a widely used marker, but its lack of specificity necessitates complementary biomarkers.
Integrated Panels:
Combining host gene expression, metabolomic data, and proteomic findings improves diagnostic accuracy.

4. Relationship Between Genetic Variations and Outcomes

Rare genetic variants, such as loss-of-function mutations, are implicated in sepsis susceptibility.
Key Examples:

**DDAH2 (Dimethylarginine Dimethylaminohydrolase 2)**: SNP in the promoter region affects nitric oxide metabolism and outcomes.

**PCSK9 (Proprotein Convertase Subtilisin/Kexin Type 9)**: Modulates lipid metabolism and cytokine production, influencing survival.

5. Mitochondrial Dysfunction and Muscle Loss

Key Points Discussed:
Overview:

Mitochondria play a critical role in energy metabolism, especially in tissues like muscles.
In septic shock survivors, mitochondrial dysfunction is a significant contributor to long-term complications, including muscle loss (myopathy).
Mechanisms of Mitochondrial Dysfunction in Sepsis:

Reactive Oxygen Species (ROS):Overproduction during sepsis damages mitochondrial DNA (mtDNA), proteins, and lipids, impairing mitochondrial function.

Mitochondrial Membrane Potential Collapse:Damage disrupts the electron transport chain, leading to decreased ATP production.

Release of mtDNA:Damaged mitochondria release mtDNA into circulation, where it acts as a damage-associated molecular pattern (DAMP), further fueling inflammation.

Impact on Muscle:Loss of mitochondrial efficiency leads to impaired energy production in skeletal muscles.
Prolonged mitochondrial dysfunction contributes to critical illness myopathy (CIM), characterized by:
Loss of muscle mass.
Reduced muscle strength.
Long-term physical disability in septic shock survivors.

Biomarkers of Mitochondrial Dysfunction:

**Cardiolipin**: A key mitochondrial membrane lipid, disrupted during sepsis.

**Circulating mtDNA**: Indicates mitochondrial injury and systemic inflammation.

**Carnitine Esters**: Reflect mitochondrial energy metabolism dysfunction.

Therapeutic Considerations:
Interventions to restore mitochondrial function, such as antioxidants or metabolic support, could mitigate muscle loss in septic patients.


#### 15.The Use of Biomarkers in the Diagnosis and Management of Sepsis
International Journal of Biomedical Laboratory Science 2020(not found)

1. Inflammatory Biomarkers
Biomarkers linked to inflammation, though non-specific for sepsis.

**Lactate**
Role: A metabolite of glucose elevated in anaerobic metabolism.
Clinical Use: Elevated lactate (>4 mmol/L) predicts higher mortality in sepsis and indicates tissue hypoxia.
Limitations: Non-specific; must be assessed with clinical context.

**C-Reactive Protein (CRP)**
Protein Role: Acute-phase protein produced by the liver in response to cytokines like IL-6.
Function: Enhances phagocytosis by binding microbial phosphocholine and damaged cells.
Clinical Use: Inflammation marker; moderately elevated in chronic conditions and peaks in infections after 12–24 hours.
Gene: CRP
High-Sensitivity CRP (hsCRP)
Similar to CRP but offers more sensitivity for low-grade inflammation.

**Procalcitonin (PCT)**
Protein Role: Precursor of calcitonin produced in bacterial infections.
Function: Differentiates bacterial from non-bacterial infections; rapidly elevated in sepsis.
Clinical Use: Guides antibiotic therapy initiation and discontinuation.
Gene: CALCA

**Heparin-Binding Protein (HBP)**
Protein Role: Found in neutrophil granules.
Function: Induces vascular leakage and promotes immune cell chemotaxis.
Clinical Use: Indicates neutrophil activation; elevated levels correlate with severe infections.

2. Immune Response Biomarkers

Markers reflecting immune dysregulation during sepsis.

**Interleukin-6 (IL-6)**
Role: Cytokine produced by macrophages, monocytes, and others during inflammation.
Function: Stimulates acute-phase protein production and B-cell differentiation.
Clinical Use: Rapidly rises during infection and drops as inflammation resolves.
Gene: IL6

**Interleukin-8 (IL-8)**
Role: Chemokine for neutrophil recruitment to inflammation sites.
Function: Enhances phagocytosis and predicts pediatric sepsis outcomes.
Clinical Use: Specialized marker; high levels correlate with organ failure risk.
Gene: CXCL8

**Interleukin-10 (IL-10)**
Role: Anti-inflammatory cytokine.
Function: Regulates T-cell cytokine production and promotes B-cell survival.
Clinical Use: Differentiates SIRS from sepsis.
Gene: IL10

**Presepsin (sCD14)**
Role: Soluble form of CD14 protein on macrophages.
Function: Early marker distinguishing bacterial sepsis from SIRS.
Clinical Use: Elevated levels predict mortality and bacterial infection severity.
Gene: CD14

**Monocyte Chemoattractant Protein-1 (MCP-1/CCL2)**
Role: Chemokine for recruiting monocytes to infection sites.
Function: Linked to inflammation and SOFA scores.
Clinical Use: Predicts severity and outcomes in sepsis patients.
Gene: CCL2

**CD64 (FcγRI)**
Role: High-affinity receptor for IgG.
Function: Promotes antigen capture and immune cell activation.
Clinical Use: Differentiates bacterial infections from other causes.
Gene: FCGR1A

**Soluble Urokinase-Type Plasminogen Activator Receptor (suPAR/CD87)**
Role: Found on immune cells and tumor cells.
Function: Elevated in bacterial infections; levels >8 ng/mL predict poor prognosis.
Clinical Use: Indicator of mortality risk.
Gene: PLAUR

**Soluble Triggering Receptor Expressed on Myeloid Cells-1 (sTREM-1)**
Role: Found on neutrophils and monocytes.
Function: Upregulated in bacterial infections but not in non-infectious inflammation.
Clinical Use: Helps distinguish sepsis from non-infectious SIRS.
Gene: TREM1

3. Application in Therapy and Prognosis
Combination Biomarker Panels
Suggested combinations: PCT, presepsin, CD64, suPAR, and sTREM-1.
Panels improve diagnostic and prognostic accuracy compared to single markers.
Procalcitonin for Antibiotic Guidance
Low PCT levels (<0.5 μg/L) indicate no need for antibiotics.
High PCT levels (>2 μg/L) necessitate immediate antibiotic initiation.
