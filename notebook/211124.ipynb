#### Found an article about Use of Gene-Expression Profiling to Identify Candidate Genes in Human Sepsis(folder gene expression level n 1.)
Key Genes Mentioned
1. NFKBIA:
Function: Encodes the inhibitor of nuclear factor-kappa B (NF-κB).
Role in Sepsis:
Significantly upregulated in septic patients.
Reflects suppression of the NF-κB pathway, a critical regulator of immune and inflammatory responses.
Suggests that sepsis inhibits immune activation through increased expression of this inhibitor.
2. MYD88:
Function: Encodes an adaptor protein involved in Toll-like receptor (TLR) and interleukin-1 receptor signaling.
Role in Sepsis:
Plays a central role in initiating inflammatory responses.
Downregulated in sepsis, highlighting reduced immune activation.
3. TNFRSF1A and TNFRSF1B:
Function: Encode members of the tumor necrosis factor receptor (TNF receptor) superfamily.
Role in Sepsis:
Reduced expression suggests a dampened inflammatory response to infection.
These receptors mediate signals for cell survival, apoptosis, and inflammation.
4. MAP3K (MAPKKK):
Function: Encodes mitogen-activated protein kinase kinase kinase, a key regulator of the MAPK signaling pathway.
Role in Sepsis:
Downregulated, reflecting suppressed inflammatory signaling.
5. RAF1:
Function: Encodes a protein that regulates the mitogen-activated protein kinase (MAPK) pathway.
Role in Sepsis:
Its reduced expression indicates suppressed inflammatory pathways during sepsis.
6. Genes Involved in Mitochondrial Function:
Examples: Genes related to oxidative phosphorylation and ATP synthesis.
Role in Sepsis:
Downregulated in septic patients.
Suggests mitochondrial dysfunction, an adaptive response to overwhelming inflammation.


#### Also found another article The Relevance of Coding Gene Polymorphisms of Cytokines and Cellular Receptors in Sepsis(folder gene expression level n 2.)
This article reviews the role of genetic polymorphisms in cytokines and cellular receptors in sepsis. It explores how genetic variations influence susceptibility, progression, and outcomes of sepsis and identifies potential biomarkers for diagnosis and prognosis.

Cytokine Gene Polymorphisms
Tumor Necrosis Factor-Alpha (TNF-α):
Polymorphisms at positions -238, -308, -376, and +489 are associated with increased TNF-α levels.
Elevated TNF-α correlates with the severity of sepsis but has variable effects on prognosis.
Interleukin-6 (IL-6):
SNP rs1800795 (G/C at -174) affects IL-6 plasma levels, influencing acute-phase protein production (e.g., CRP).
Evidence on the association between this SNP and sepsis risk is inconsistent.
Interleukin-10 (IL-10):
SNP -1082 G/A affects IL-10 production and sepsis outcomes.
High IL-10 levels correlate with severe inflammation and higher mortality in sepsis.
Cellular Receptor Gene Polymorphisms
Toll-Like Receptor 4 (TLR4):
SNPs like +896 (rs4986790) and Asp299Gly alter TLR4 expression, impacting susceptibility to sepsis and septic shock.
Cluster of Differentiation 14 (CD14):
Polymorphism -159 C/T is linked to sepsis risk, but findings are conflicting.
CD14's soluble form (presepsin) is a sensitive diagnostic biomarker for sepsis.
Triggering Receptor Expressed on Myeloid Cells-1 (TREM-1):
Variants like rs223426 increase susceptibility to septic shock but do not consistently affect soluble TREM-1 levels.
Other Biomarkers
Plasminogen Activator Inhibitor-1 (PAI-1):
Polymorphism 4G/5G is associated with increased sepsis mortality.
Soluble Urokinase-Type Plasminogen Activator Receptor (suPAR):
Elevated suPAR levels correlate with sepsis severity but vary across studies.

Previously identified six miRNAs linked to sepsis outcomes:
miR-223, miR-15a, miR-16, miR-122, miR-193b, and miR-483-5p.
