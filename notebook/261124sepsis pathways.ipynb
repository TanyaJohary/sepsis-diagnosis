Sepsis pathways:
## 1. Inflammatory and Immune Pathways

Cytokine Storm: Overproduction of pro-inflammatory cytokines (e.g., TNF-α, IL-1β, IL-6) leads to systemic inflammation.**TNF, IL1A, IL1B, IL6, IL8 (CXCL8), IL12A, IL12B, IFNG, TREM-1, TYROBP, LBP, HMGB1, ELANENeutrophil Extracellular Traps (NETs), CXCL8**

Anti-inflammatory cytokines (e.g., IL-10, TGF-β) are also upregulated but may result in immunosuppression in later stages of sepsis.**IL10, TGFB1, IL1RN (IL-1 receptor antagonist), SOCS1, SOCS3**


Toll-like Receptor (TLR) Signaling:TLRs recognize pathogen-associated molecular patterns (PAMPs) and activate immune responses.Dysregulated TLR signaling exacerbates inflammation and tissue damage.
**TLR1, TLR2, TLR3, TLR4, TLR5, TLR7, TLR8, TLR9, MYD88, TICAM1 (TRIF), IRAK1, IRAK4, TRAF6, TIRAP, LBP**

Inflammasome Components: **CASP1 (Caspase-1), GSDMD, PYCARD (ASC), AIM2**

NOD-like Receptors:  **NLRP3, NLRC4, NLRP1, NOD1, NOD2, RIPK2**

Complement Pathway Activation:Excessive complement activation contributes to inflammation, vascular permeability, and tissue injury.
**C3, C3aR1, C5, C5AR1 (C5a receptor), CFB, CFD, CFH, C1QA, C1QB, C1QC**

## 2.Acute Phase Response Pathway

Pathophysiological Perspective:

The acute phase response is triggered by inflammatory cytokines such as IL-6, TNF-α, and IL-1β during infection or tissue injury.

It involves the rapid production of acute-phase proteins like C-reactive protein (CRP), serum amyloid A (SAA), and fibrinogen by the liver.

These proteins amplify inflammation, modulate the immune response, and facilitate pathogen clearance.

Molecular Perspective:Key Genes:

Acute Phase Proteins: **CRP, SAA1, HP (Haptoglobin), FGA, FGB, FGG (Fibrinogen chains), LBP**

Cytokines Triggering the Acute Phase Response: **IL6, IL1B, TNF**

Liver Detoxification Genes (Supporting Role): **CYP3A4, CYP2C9, CYP1A2**

## 3. Coagulation Pathways

Pathophysiological Perspective:

Disseminated Intravascular Coagulation (DIC):Excessive clot formation depletes clotting factors, leading to microvascular thrombosis and bleeding.

Tissue Factor (TF) Pathway Activation:Overexpression of TF leads to thrombin generation, fibrin deposition, and clot formation.

Protein C Pathway Dysfunction:Impaired Protein C activation leads to loss of anticoagulant and anti-inflammatory effects.

Fibrinolytic Impairment:Increased PAI-1 expression inhibits plasminogen activation, reducing clot breakdown.

Molecular Perspective:
Key Genes:

Coagulation factors: **F2 (Prothrombin), F3 (Tissue Factor), F5, F7, F8, F9, F10, F11, ELANE , CXCL8**

Anticoagulation pathway: **PROC (Protein C), PROS1 (Protein S), SERPINA5 (Protein C Inhibitor), THBD (Thrombomodulin)**

Fibrinolytic pathway: **PLG (Plasminogen), PLAU (Plasminogen Activator), SERPINE1 (PAI-1), PLAUR**

## 4. Vascular and Endothelial Pathways

Pathophysiological Perspective:

Endothelial Dysfunction:Endothelial cells lose integrity, causing vascular leakage and tissue edema.

Microvascular Thrombosis:Microvascular clots impair blood flow, leading to organ ischemia.

Vasodilation and Hypotension:Excess nitric oxide (NO) causes vasodilation, leading to septic shock.

Molecular Perspective:
Key Genes:

Endothelial dysfunction: **SELE, ICAM1, VCAM1, PECAM1, VEGFA, TREM1, HMGB1, ELANE, CXCL8** 

Vascular barrier proteins: **CLDN5, OCLN, MMP9, MMP2**

Nitric oxide regulation: **NOS2 (iNOS), NOS3 (eNOS), ARG1**

Angiopoietin-Tie2 signaling: **ANGPT1, ANGPT2, TEK (Tie2)**

## 5. Metabolic Pathways

Pathophysiological Perspective:

Hypermetabolism:Increased energy demand causes hyperglycemia and lipolysis.

Mitochondrial Dysfunction:Impaired ATP production leads to energy deficits.

Lactic Acidosis:Hypoperfusion and anaerobic glycolysis cause lactate buildup.

Molecular Perspective:
Key Genes:

Glucose metabolism: **GCK, HK2, SLC2A4, PDK1, SLC2A4 (GLUT4), INS (Insulin), INSR (Insulin Receptor)**

Mitochondrial dysfunction: **MT-ND1, MT-ND5, ATP5F1A, SDHB**

Lactate metabolism:**LDHA, LDHB,  PDK1, PDK4**

Lipid Metabolism:**CPT1A, PPARG, SREBF1 (SREBP1), FABP4 (Fatty Acid Binding Protein)**


## 6. Oxidative Stress Pathways

Pathophysiological Perspective:

Reactive Oxygen Species (ROS) Overproduction:ROS damage lipids, proteins, and DNA, leading to cell injury.

Antioxidant Depletion:Insufficient antioxidant defenses exacerbate oxidative damage.

Molecular Perspective:Key Genes:

ROS generation: **NOX1, NOX2 (CYBB), NOX4, SOD1, SOD2, TREM1, ELANE**

Antioxidant defense: **GPX1, GPX4, CAT, PRDX1**

## 7. Apoptosis and Cell Death Pathways

Pathophysiological Perspective:

Lymphocyte Apoptosis:Immune cell death weakens host defense.

Endothelial Cell Death:Apoptosis and pyroptosis disrupt vascular integrity.

Molecular Perspective:
Key Genes:

Caspase family: **CASP3,CASP6, CASP7, CASP8, CASP9**

Pro-apoptotic genes: **BAX, BAK1**

Anti-apoptotic genes: **BCL2,BCL2L1 (Bcl-xL), MCL1**

Pyroptosis regulators: **NLRP3, CASP1, GSDMD, PYCARD, AIM2, HMGB1**

## 8. Neuroendocrine Pathways

Pathophysiological Perspective:

Hypothalamic-Pituitary-Adrenal (HPA) Axis Dysregulation:Abnormal cortisol production affects immune and metabolic responses.

Sympathetic Nervous System Activation:Excess catecholamines drive hypermetabolism and cardiovascular stress.

Molecular Perspective:
Key Genes:

HPA axis: **CRH, POMC, NR3C1, CALCA**

Sympathetic nervous system: **ADRB1, ADRB2, TH**

## 9. Autophagy Pathways

Pathophysiological Perspective:

Autophagy Impairment:Failure to clear damaged organelles and pathogens exacerbates cell dysfunction.

Molecular Perspective:
Key Genes:

Core autophagy machinery: **ATG5, ATG7, BECN1, MAP1LC3B, SQSTM1, ULK1, SQSTM1 (p62)**


## 10. Organ-Specific Pathways

Pathophysiological Perspective:

Acute Kidney Injury (AKI):Hypoperfusion, inflammation, and tubular damage impair renal function.

Sepsis-Associated Encephalopathy (SAE):Neuroinflammation and blood-brain barrier disruption cause cognitive dysfunction.

Liver Dysfunction:Reduced detoxification and bile production contribute to metabolic dysregulation.

Cardiac Dysfunction:Myocardial depression and impaired contractility lead to shock.

Molecular Perspective:
Key Genes:

### Kidney injury:

Acute Kidney Injury (AKI):**LCN2 (NGAL), HAVCR1 (KIM-1), SOD1, UMOD (Uromodulin)**

RAAS Dysregulation: **AGT, REN, ACE, ACE2, AGTR1 (Angiotensin Receptor 1)**
            
### Brain pathways: 

Blood-Brain Barrier Integrity: **CLDN5, OCLN, MMP9, MMP2**

Neuroinflammation: **GFAP (Glial Fibrillary Acidic Protein), TNF, IL1B, IL6**

Neurotransmitter Regulation: **SLC6A3 (Dopamine Transporter), SLC6A4 (Serotonin Transporter)**

### Liver dysfunction:

Acute Phase Response:**CRP, SAA1, HP, FGA (Fibrinogen Alpha Chain), FGB, FGG**

Detoxification: **CYP3A4, CYP2C9, CYP1A2**

Cardiac dysfunction: **TNF, NOS3, MYH6,  MYH7, ACTC1 (Cardiac Actin)**

## 11. Epigenetic and Transcriptomic Regulation

Pathophysiological Perspective:

Epigenetic Changes:Histone modifications and DNA methylation regulate inflammation and immune responses.

Transcriptomic Dysregulation:Sepsis alters gene expression patterns, promoting maladaptive responses.

Molecular Perspective:
Key Genes:

Epigenetic regulators: **HDAC1, DNMT1**

Inflammation transcription factors:**NFKB1, STAT3, HIF1A**
